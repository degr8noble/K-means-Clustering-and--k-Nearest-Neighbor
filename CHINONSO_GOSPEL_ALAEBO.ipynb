# First model with y1="Result_collect"  with 11 features

```python
#to divide our sample into test and sample
from sklearn.model_selection import train_test_split
x=x3
y=y1
x_train,x_test,y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=123)
#x_train,x_test,y_train, y_test=train_test_split(feat,pred,test_size=0.3,random_state=123)
x_train.shape,x_test.shape,y_train.shape, y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
classifier1=Sequential()
classifier1.add(Dense(10,kernel_initializer="uniform", activation="relu",input_shape=(11,)))
classifier1.add(Dropout(rate=0.001))
classifier1.add(Dense(10,kernel_initializer="uniform", activation="relu",))
classifier1.add(Dense(1,kernel_initializer="uniform", activation="softmax"))
```

```python
classifier1.compile(optimizer= "adam", loss = "binary_crossentropy",metrics = ["accuracy"])
```

```python
classifier1.fit(x_train, y_train,batch_size = 100, epochs = 200,verbose = 1)
```

```python
y_pred=classifier1.predict(x_test)
```

```python
loss, score = classifier1.evaluate(x_test, y_test)
print("score:",score)
```
